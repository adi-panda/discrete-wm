# Interactive World Model Play

The video auto-plays at your chosen FPS while ▶ Play is on. Click an action button to change the *held* action (LEFT/NOOP/RIGHT stick until you change them). FIRE is one-shot — it fires for the next frame, then reverts. Adjust `FPS` or `gen steps` on the fly.

```python
import sys
sys.path.insert(0, '/root/research/discrete_wm')

from play_interactive import InteractiveWorldModel

wm = InteractiveWorldModel(
    checkpoint_path='/vast/adi/discrete_wm/checkpoints/model_best.pt',
    tokenizer_path='/vast/adi/discrete_wm/checkpoints/tokenizer_final.pt',
    data_path='/vast/adi/discrete_wm/data/tokenized_data.npz',
    gen_steps=8,
    temperature=0.9,
)
```

```python
# Live video playback.
# ▶ Play starts the auto-stepping video; click action buttons to change direction.
wm.run_live(fps=10, scale=4)
```

## Manual stepping
If you prefer more control, you can step manually and inspect frames:

```python
import matplotlib.pyplot as plt

frame = wm.reset()

# Run a sequence of actions
actions = [1, 0, 0, 2, 2, 2, 3, 3, 3, 0]  # FIRE, then move around
frames = [frame]
for a in actions:
    frames.append(wm.step(a))

# Display as a grid
fig, axes = plt.subplots(1, len(frames), figsize=(2 * len(frames), 2))
names = ['start'] + [['NOOP','FIRE','RIGHT','LEFT'][a] for a in actions]
for ax, f, n in zip(axes, frames, names):
    ax.imshow(f)
    ax.set_title(n, fontsize=8)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
